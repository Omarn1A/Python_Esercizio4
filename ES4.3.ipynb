```python
diabetici_minori20 = df['Age (years)'] <= 20

# Maschera per diabetici con età compresa tra 20 e 30
diabetici_compresi2030 = (df['Age (years)'] >= 20) & (df['Age (years)'] <= 30)

# Maschera per diabetici con età compresa tra 30 e 40
diabetici_compresi3040 = (df['Age (years)'] >= 30) & (df['Age (years)'] <= 40)

# Maschera per diabetici con età compresa tra 40 e 50
diabetici_compresi4050 = (df['Age (years)'] >= 40) & (df['Age (years)'] <= 50)

# Maschera per diabetici con età >= 50
diabetici_maggiori50 = df['Age (years)'] >= 50
```

```python
df_minori20=df.loc[diabetici_minori20]
df_minori20.describe()
```

```python
df_tra2030=df.loc[diabetici_compresi2030]
df_tra2030.describe()
```

```python
df_tra3040=df.loc[diabetici_compresi3040]
df_tra3040.describe()
```

```python
df_tra4050=df.loc[diabetici_compresi4050]
df_tra4050.describe()
```

```python
df_oltre50=df.loc[diabetici_maggiori50]
df_oltre50.describe()
```
